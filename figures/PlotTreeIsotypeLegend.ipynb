# Plot legend for phylogenies

```python
import json

%matplotlib inline
from matplotlib import pyplot as plt
import matplotlib as mpl
import seaborn as sns

output_dir = "/local10G/rfhorns/Bcell/flu_highres/figures/v5/outs"
output_suffix = ""
output_formats = [".png", ".pdf"]

def save_figure(fig, name, output_dir, output_suffix, output_formats, savefig_args):
    for output_format in output_formats:
        fig.savefig(output_dir + "/" + name + output_suffix + output_format, **savefig_args)
    return None

savefig = True
savefig_args = {"dpi": 300, "bbox_inches": "tight", "pad_inches": 0.2}
mpl.rc('savefig', dpi=300)

sns.set_style("ticks")
sns.set_context("talk")

myColors = ["#E69F00", "#56B4E9"]
myColors_subjects = ['#e41a1c','#377eb8','#a6cee3','#984ea3','#f781bf']
```

```python
# Load colors
infile = "isotype_to_color_dict.json"
with open(infile) as f:
    isotype_to_color_dict = json.load(f)
```

```python
# Plot as scatter

isotypes_ordered = ["IgM", "IgD", "IgA1", "IgA2", "IgE",
                    "IgG1", "IgG2", "IgG3", "IgG4", "germline"]

fig, ax = plt.subplots(1, 1, figsize=(6,4))

for isotype in isotypes_ordered:
    color = isotype_to_color_dict[isotype]
    ax.scatter(0, 0, c=color, label=isotype, lw=0, s=50)

ax.set_xlim(1,2)
ax.legend(loc="center", ncol=2, title="Isotype")
plt.axis('off')
plt.tight_layout()

save_figure(fig, "IsotypeLegend", output_dir, output_suffix, output_formats, savefig_args)
```
